```python
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt

import plotly.express as px
```

```python
twm_customers = pd.read_csv("data/twm_customer.csv", sep=";")
```

```python
twm_customers['state_code'] = twm_customers['state_code'].map(str.strip)
twm_customers['region'] = twm_customers['state_code']
#1 : West
#2 : Southwest
#3 : Midwest
#4 : Northeast
#5 : Southeast
twm_customers = twm_customers.replace(
    {'region': {'IL': 3, 'IN': 3, 'IA': 3, 'KS': 3, 'MI': 3, 'MN': 3, 'MO': 3, 'NB': 3, 'ND': 3, 'OH': 3,
                'SD': 3, 'WI': 3,
                'CT': 4, 'DE': 4, 'ME': 4, 'MD': 4, 'MA': 4, 'NH': 4, 'NJ': 4, 'NY': 4, 'PA': 4, 'RI': 4,
                'VT': 4, 'DC': 4,
                'AL': 5, 'AR': 5, 'FL': 5, 'GA': 5, 'KY': 5, 'LA': 5, 'MS': 5, 'NC': 5, 'SC': 5, 'TN': 5,
                'AZ': 2, 'NM': 2, 'OK': 2, 'TX': 2, 'VA': 2, 'WV': 2,
                'AK': 1, 'CA': 1, 'CO': 1, 'HI': 1, 'ID': 1, 'MT': 1, 'NV': 1, 'OR': 1, 'UT': 1, 'WA': 1,
                'WY': 1}})

```

```python
def dummy_var_create(df, column_name):
    for name in set(df[column_name]):
        if str(name).strip() != "":
            df[f"{column_name}_dummy_{name}"] = (df[column_name] == name).astype(int)


dummy_var_create(twm_customers, 'region')
dummy_var_create(twm_customers, 'marital_status')
twm_customers['is_male'] = (twm_customers['gender'].map(str.strip) == 'M').astype(int)

twm_customers = pd.get_dummies(twm_customers, prefix='region', columns=['region'])
twm_customers.rename(
    columns={'region_1': 'isWest', 'region_2': 'isSouthwest', 'region_3': 'isMidwest', 'region_4': 'isNortheast',
             'region_5': 'isSoutheast'})
```

```python
result = twm_customers[twm_customers['state_code'] == "NY"]
print(result)
```

```python
for i in set(twm_customers):
    if len(set(twm_customers[i])) <= 50:
        print(f"{i} : {set(twm_customers[i])}")
```

```python
twm_customers2 = pd.DataFrame(twm_customers)
twm_customers2.marital_status = twm_customers.marital_status.astype(str)
twm_customers2.age = (twm_customers.age // 10) * 10
first_scatter = px.scatter(twm_customers2, "age", "income", size="years_with_bank", color="marital_status",
                           title="Size of bubble represents number of years with bank")
first_scatter
```
